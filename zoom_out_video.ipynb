```python
import os, sys
import cv2
import matplotlib.pyplot as plt
```

```python

img = cv2.imread("images/test.jpg")
print(img.shape)
plt.imshow(img)
```

```python
n_grid = 21
horizontal_images = cv2.hconcat([img]*n_grid)
full_image = cv2.vconcat([horizontal_images]*n_grid)
print(full_image.shape)
```

```python
plt.imshow(full_image) # takes time to load very big image          
```

```python
# Get the center of the image
center_y, center_x, channels = full_image.shape
center_x = center_x // 2
center_y = center_y // 2
print(center_x, center_y, full_image.shape)
```

```python
h_out , w_out, _ = img.shape
h_start = h_out//2
w_start = w_out//2


fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video_name = 'output.mp4'
fps = 1
video = cv2.VideoWriter(video_name, fourcc, fps, (w_out, h_out))

iter = 9 # Make sure this isnt doesnt lead to more that the actual width of the full image

while(iter):
    temp_image = full_image[center_y-h_start:center_y+h_start, center_x-w_start:center_x+w_start]
    temp_image = cv2.resize(temp_image, (w_out, h_out))
    video.write(temp_image)
    
    h_start+=h_out
    w_start+=w_out
    iter=iter-1
```
